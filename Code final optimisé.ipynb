```python
import urllib.request #chargement de la librarie REQUESTS
```

```python
from bs4 import BeautifulSoup # Chargement de la librairie BeautifulSoup
```

```python
import pandas as pd # Chargement de la librairie Pandas
```

```python
import validators 
# Chargement de la librairie Validators
# Parmi les fonctions de cette librairie Validators.url(lien,Public=False) qui fait un test si le lien est valable ou non.
```

```python
df = pd.read_excel(r'atelier_stat.xlsx',header=0) # Lecture du fichier excel en un dataframe
```

```python
l= [x for x in df['custumer name']] # Conversion en liste pour une manipulation plus facile
```

```python
for i in l:
    print(i)
# affichage des Noms des Clients
```

```python
result = pd.DataFrame(columns=['Customer Name','Website','Facebook','Instagram','LinkedIn']) #création d'un nouveau dataframe
```

```python
for i in l : #le parcours de la liste pour scraper le contenu
    
    new_row= {"Customer Name":"","Website":"","Facebook":"","Instagram":"","LinkedIn":""} #création d'un dictionnaire afin d'ajouter une nouvelle ligne
    
    new_row["Customer Name"]=i #stockage du nom de clients
    
    url = 'https://google.com/search?q='+urllib.parse.quote_plus(i) #création d'un lien 
    
    request = urllib.request.Request(url) #demande au serveur de google pour avoir la page HTML
    
    request.add_header('User-Agent', 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36')
    
    raw_response = urllib.request.urlopen(request).read() # Lecture de la page HTML 
    
    html = raw_response.decode("utf-8") #décodage selon UTF-8
    
    soup = BeautifulSoup(html, 'html.parser') # Parser la page HTML
    
    wb=soup.find('link',hreéf=True) # Extraction du lien envers le site 
    
    if (wb!=None and validators.url(wb['href'], public=False)): #controle et stockage dans la nouvelle ligne
        new_row["Website"]=wb['href']
        
        fb=soup.select_one('a[href*=facebook]') #extraction du lien facebook
        if(fb!=None): #controle et stockage dans la nouvelle ligne
            new_row["Facebook"]=fb['href']
        else: new_row["Facebook"]="Page Facebook Introuvable"

        ig=soup.select_one('a[href*=instagram]') #extraction du lien instagram
        if(ig!=None): #controle et stockage dans la nouvelle ligne
            new_row["Instagram"]=ig['href']
        else: new_row["Instagram"]="Profil Instagram Introuvable"

        li=soup.select_one('a[href*=linkedin]')  #extraction du lien LinkedIn
        if(li!=None):  #controle et stockage dans la nouvelle ligne
            new_row["LinkedIn"]=li['href']
        else:new_row["LinkedIn"]="Profil LinkedIn introuvable"
    else:
        continue
    result = result.append(new_row, ignore_index=True) # Ajout de la nouvelle ligne au dataframe
    
```

```python
l = [x for x in result["Website"]] #création d'une liste contenant les liens des sites officiels
```

```python
for i in l : print(i) #affichage des liens
```

```python
#la recherche des logos 
s=0
links=[]
for i in l:
    request = urllib.request.Request(i)
    
    request.add_header('User-Agent', 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36')
    
    response = urllib.request.urlopen(request).read()
    
    html=response.decode("utf-8")
    
    soup = BeautifulSoup(html, 'html.parser')
    
    tags=soup.select_one('img[src*=logo]')
    
    x=soup.find('img',src=True)
    if (tags==None and x==None):
        links.append('No Logo')
    
    else:
        if(tags!=None):
            links.append(tags)
            print(links[s])
            s=s+1  
    
        else:
            if(x!=None and tags==None):
                links.append(x)
                print(links[s])
                s=s+1
                
```

```python
# éxtraction des liens sources des logos et leurs stockage dans une liste
# l'affiche de chaque étape de l'éxecution est facultatif mais sert à mieux expliquer le code.
logo_src=[]
s=0
for i in links:
    if (i!='No Logo'):
        x=i['src']
        if (l[s])in x :
            print(s+1,x) 
            logo_src.append(x)
            s=s+1
        elif ((('http://' in x) or ('https://' in x) )  and (l[s] not in x)):
            print(s+1,x)
            logo_src.append(x)
            s=s+1
        elif (x[0]=='/' and l[s] not in x):
            print(s+1,l[s]+x[1:])
            logo_src.append(l[s]+x[1:])
            s=s+1
        elif (x[:1]=='//' and l[s] in x):
            print(s+1,x[2:])
            logo_src.append(x[2:])
            s=s+1
        else:
            print(s+1,l[s]+x)
            logo_src.append(l[s]+x)
            s=s+1
            
    else:
        logo_src.append(i)
```

```python
result.insert(5, "Logo Source", logo_src, True) 
#ajout d'une nouvelle colonne dans la table 
```
